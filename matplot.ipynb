```python
from matplotlib import style
style.use("ggplot")
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'g',label='line one',linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)

plt.title("Epic info")
plt.ylabel("Y-axis")
plt.xlabel("X-axis")

plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
slices=[1,2,3,4]
activities=["sleep","work","eat","slice"]
cols=['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,startangle=180,shadow=True,explode=(0,0.2,0,0),autopct="%1.1f%%")
plt.title("piechart")
plt.show()
```
